# Training

```python
x=np.array(df['DEPARTURE_DELAY']).reshape(-1,1)
y=np.array(df['ARRIVAL_DELAY']).reshape(-1,1)
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train.shape
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
model=reg.fit(x_train,y_train)
print(model.coef_)
print(model.intercept_)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn import metrics
mean_abs_error=metrics.mean_absolute_error(y_pred,y_test)
mean_sqr_error=metrics.mean_squared_error(y_pred,y_test)
root_mean_error=np.sqrt(metrics.mean_squared_error(y_pred,y_test))
print(mean_abs_error)
print(mean_sqr_error)
print(root_mean_error)
```

```python
y_mean=df['ARRIVAL_DELAY'].mean()
squared_diff=(df['ARRIVAL_DELAY']-y_mean)**0.5
SST=squared_diff.sum()
R=(1-mean_sqr_error/SST)**0.5
print(R)
```

# GRAPH

```python
plt.scatter(x_test,y_test,c='r',label='Actual')
plt.scatter(y_pred,y_test,c='g',label='Predicted')
plt.legend()
plt.show()
```
